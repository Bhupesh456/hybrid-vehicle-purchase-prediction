```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(r"training_data.csv")

data.head()
```

```python
x = data[["Age","Education","Income(Lacs)","Vehicle Ownership"]]
y = data["Purchase hybrid vehicle"]
```

```python
model = LinearRegression()

model.fit(x,y)

print("The Model is trained and ready to make predictions")
```

```python
new_data = pd.read_csv(r"new_customer_data.csv")

new_data.head()
```

```python
new_data["Predicted Purchase % "] = ((model.predict(new_data))*100).round(2)

new_data.head()
```

```python
new_data.to_csv("Car_Prediction.csv", index = False)
```

Visulaization


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.histplot(new_data["Predicted Purchase % "],bins=20,kde=True)

plt.title("Distribution of Predicted Purchase Percentage")
plt.xlabel("Predicted Purchase (%)")
plt.ylabel("Number of Customers")

plt.grid(alpha=0.3)

plt.savefig("prediction_distribution.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(new_data["Income(Lacs)"],new_data["Predicted Purchase % "],alpha=0.7)

plt.xlabel("Income (Lacs)")
plt.ylabel("Predicted Purchase (%)")
plt.title("Income vs Predicted Purchase")

plt.grid(alpha=0.3)

plt.savefig("income_vs_prediction.png", dpi=300, bbox_inches="tight")

plt.show()
```
